**KEBERADAAN MISSING VALUES**

> 2b) Dampak Missing Values terhadap Analisis Data
      Missing Values ini secara sederhana dapat diartikan sebagai adanya data yang kosong dalam database. Apabila muncul adanya missing values ini, akan berakibat terhadap analisis data yang memungkinkan adanya bias. Dalam imputasi terdapat tiga cara yang dilakukan yaitu mean (data berdistribusi normal), median (dipakai ketika ada outlier), dan modus (ketika data berbentuk kategorik atau diskrit).




HANDLING MISSING VALUE

> Dalam Machine Learning, missing values adalah masalah umum yang harus ditangani sebelum model dapat digunakan. Ada beberapa metode untuk menangani missing values, tergantung pada jenis data dan karakteristik distribusinya.



2c) Metode penanganan missing values
> 1. Penanganan untuk kolom harga
      Harga merupakan data yang menunjukkan data yang kontinu dan biasanya berdistribusi normal atau mendekati normal. Lalu apabila terdapat missing values di kolom ini, metode yang paling sesuai adalah menggunakan imputasi mean. Alasannya jika missing values digantikan dengan rata-rata harga database maka tidak akan terlalu memengaruhi model. Rata-rata harga akan mewakili kecenderungan data.

> 2. Penanganan untuk kolom rating

Menggunakan modus karena nilai yang sering muncul dapat memastikan data yang diisi ini menggambarkan mayoritas rating dari pengguna sendiri sehingga akan mengakibatkan pola distribusinya sendiri tidak berubah signifikan.

> 3. Penanganan untuk kolom jumlah_ulasan
      Jumlah ulasan sama halnya dengan harga karena datanya dapat menggunakan data yang kontinu. oleh karena itu metode yang paling tepat yaitu imputasi dengan mean.

**2. IQR**

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile (0.25) #Kuartil 1
    Q3 = data[column].quantile (0.75) #Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

**PENANGANAN OUTLIER**

> Metode penanganan outlier ada dua yaitu Winsorizing dan penghapusan

> 1. Winsorzing dapat dilakukan dengan cara menggantikan outlier dengan batas IQR nya

> 2. Penghapusan dilakukan apabila outlier jumlahnya sedikit dan tidak signifikan


**3c) Metode yang sesuai**
> Berdasarkan dataset yang ada, diperoleh output bahwa terdapat outlier pada kolom harga sebanyak 14 outliers. Penanganan outlier yang sesuai untuk kasus tersebut adalah menggunakan metode Winsorizing. Alasannya dikarenakan metode ini memperhatikan IQR nya, dimana mempuanya batas-batas tenrtentu. melalui metode ini, nantinya outlier akan digantikan dengan batas IQR tersebut sehingga nilai yang tinggi atau nilai yang rendah akan diganti dengan batas yang telah ditentukan. Hal tersebut akan mendapatkan hasil analisis yang lebih akurat tanpa perlu adanya pengurangan atau bahkan penghilangan informasi penting.

```python
def winsorize_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])

# 3d) Penanganan outlier dengan menerapkan Winsorizing
for col in num_cols:
    winsorize_iqr(data, col)
```

**ENCODING DATA**

> Encoding adalah proses mengubah data kategorik menjadi bentuk numerik agar dapat digunakan dalam algoritma Machine Learning. Model Machine Learning umumnya hanya dapat bekerja dengan angka, sehingga data kategorik harus dikonversi terlebih dahulu sebelum digunakan dalam pemodelan.


**METODE ENCODING DATA**

> Ada beberapa metode encoding yang umum digunakan, tergantung pada jenis dan karakteristik data kategorik:


1.   One-Hot Encoding (OHE): Mengubah kategori menjadi beberapa kolom biner (0/1). Digunakan jika jumlah kategori tidak terlalu banyak (menghindari dimensionalitas tinggi) hal ini akan menambah dimensi kolom data.
2.   Ordinal Encoding : Mengubah kategori menjadi angka berurutan (0,1,2,..). Digunakan jika kategori memiliki urutan yang jelas (misalnya pendidikan: SD < SMP < SMA)
3.   Target Encoding : Menggunakan rata-rata target (misalnya rata-rata skor IMDB berdasarkan negara) untuk mengganti kategori. Digunakan jika data memiliki hubungan statistik yang kuat dengan target.
